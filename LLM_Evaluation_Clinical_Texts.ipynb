```python
# Testes
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer("all-MiniLM-L6-v2")
model.max_seq_length = 512

print("Max Sequence Length:", model.max_seq_length)

sentence1 = 'O animal fez barulho'

# Our sentences to encode
sentences = [
    "Que dia lindo",
    "o cachorro latiu",
    "dia bonito.",
    "O ANIMAL FEZ BARULHO"
]

# Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)
embedding1 = model.encode(sentence1, convert_to_tensor=True)

# Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    #print("Embedding:", embedding)
    cosine_scores = util.cos_sim(embedding1, embedding)
    print('cosine_scores:', cosine_scores)
    #print("")
```

```python
# Função utilizada para colocar a abreviação de uma palavra caso ela exista
def colocarAbreviacao(root, dado, abreviacao, significado):
  dadoAberto = ""
  # Vamos ver os valores 1 por 1
  # A abreviação existir, vamos adicionar o significado dela no dado
  for index, valor in abreviacao.iloc[:, 0].items():
    if str(valor).lower() == dado:
      dadoAberto = dado + "(" + str(significado.iloc[index, 0]).lower() + ")"
      return dadoAberto
  # Caso o dado não possua abreviação, devolva o dado original
  return dado

# Função para retornar um dicionário com todas as relações, incluindo o tipo de relacionamento
def relacoes(root):
    relacaoDicionario = {}
    # Mostrando as relações
    for rel in root.find('RELATIONS'):
        # O primeiro id
        an1 = rel.get('annotation1')
        # O segundo id
        an2 = rel.get('annotation2')
        # O tipo de relacionamento
        tipo = rel.get('reltype')
        # Se já houver uma entrada para este ID, adicionamos esta nova relação a uma lista
        if an1 in relacaoDicionario:
            relacaoDicionario[an1].append({'id_relacionado': an2, 'tipo_relacionamento': tipo})
        else:
            # Caso contrário, criamos uma nova entrada no dicionário
            relacaoDicionario[an1] = [{'id_relacionado': an2, 'tipo_relacionamento': tipo}]
    return relacaoDicionario

# Função usada caso exista uma negação de sintoma
def dadosRelacionados(dicionarioRelacao, id, root, dado):
    # Variáveis para retornar
    dadoFinal = ""
    verNegado = False
    # Para cada item relacionado, criamos um dado com a relação
    for key, value_list in dicionarioRelacao.items():
        for value in value_list:
            if id == value['id_relacionado']:
                idRelacao = root.find(f".//annotation[@id='{key}']")
                if idRelacao is not None:
                    relacao = idRelacao.get('text').lower() + " "
                    dadoFinal += relacao
                    # Verificando se o tipo de relacionamento é uma negação
                    if value['tipo_relacionamento'] == 'negation_of':
                        verNegado = True
    dadoFinal += dado
    return dadoFinal, verNegado
```

```python
def removerBarras(dado):
    return dado.strip('|')

def separarDados(Texto):
    # Cada dado encontrado
    sintomas = []
    sindrome = []
    dadosNegados = []

    # Separando o texto em cada parte
    Partes = Texto.split("\n")
    # Encontrando cada parte e deixando no seu respectivo lugar
    for linha in range(len(Partes)):
        topico = Partes[linha].split(":")
        if len(topico) > 1:
          descricao = topico[1]
          if '|' in descricao:
            descricao = topico[1].split('|')
          else:
            descricao = topico[1].split(',')
          if "Sintomas" in topico[0]:
              sintomas.extend(descricao)
          elif "Síndromes" in topico[0] or "Sindromes" in topico[0]:
              sindrome.extend(descricao)
          elif "Dados Negados" in topico[0]:
              dadosNegados.extend(descricao)

    # Fazemos uma limpeza para cada dado de cada lista
    sintomas = [removerBarras(n.strip()) for n in sintomas if n.strip() and n.strip() != ',']
    sindrome = [removerBarras(n.strip()) for n in sindrome if n.strip() and n.strip() != ',']
    dadosNegados = [removerBarras(n.strip()) for n in dadosNegados if n.strip() and n.strip() != ',']

    # Retornamos os dados
    return sintomas, sindrome, dadosNegados
```

```python
def removerDados(sintomas, sindromes, negado):
  # Verificamos a lista de dados Negados e removemos tais dados dos sintomas, sindromes e diagnostico
  sintomas = [n for n in sintomas if n not in negado]
  sindromes = [n for n in sindromes if n not in negado]

  # Removemos dados vazios como ""
  sintomas = [n for n in sintomas if n != ""]
  sindromes = [n for n in sindromes if n != ""]

  return sintomas, sindromes
```

```python
def removerDadosVazios(dados):
    # Função para remover "**" de uma sublista
    def limparSublista(sublista):
        return [item for item in sublista if item != "**"]

    # Aplicar a função de limpeza para cada sublista em dados
    dadosLimpos = [limparSublista(sublista) for sublista in dados if sublista]
    return dadosLimpos
```

```python
def compararDados(parteNarrativa, parteModelo):
  # Variavel para segurar o que está certo
  acertos = []

  if not parteNarrativa or not parteModelo:
    return acertos, parteNarrativa, parteModelo

  # Os embeddings para mexer com o transformer
  # Sentences are encoded by calling model.encode()
  embeddingsNarrativa = model.encode(parteNarrativa)
  embeddingsModelo = model.encode(parteModelo)

  # Por enquanto vamos fazer a comparação de uma maneira simples
  # Comparamos se o sintoma da narrativa está no sintomas do modelo, mesma coisa para síndrome e diagnóstico
  index = 0
  while index < len(parteNarrativa):
    if index < 0:
      return acertos, parteNarrativa, parteModelo
    termoTexto = parteNarrativa[index]
    #print(termoTexto, "-------------")
    embeddingNarrativa = embeddingsNarrativa[index]
    if "(" in termoTexto and ")" in termoTexto:
      # Caso exista uma sigla, procuramos pela sigla em si e o significado dela
      # Ex: has e hipertensão arterial sistêmica
      termos = termoTexto.split('(')
      termos.append(termoTexto)
      termos = [n.strip(")") for n in termos]
      indexTermos = 0
      # Vemos cada parte do termo
      embeddingsTermos = model.encode(termos)
      for termo in termos:
        embeddingTermo = embeddingsTermos[indexTermos]
        for j, termoModelo in enumerate(parteModelo):
            embeddingModelo = embeddingsModelo[j]
            similaridade = util.cos_sim(embeddingTermo, embeddingModelo)
            # Vamos ser um pouco mais rigorosos na parte das siglas
            if similaridade > 0.8:
              acertos.append(termoModelo)
              # Remove os termos
              parteNarrativa.pop(index)
              parteModelo.pop(j)
              # Removemos os termos do embedding
              embeddingsNarrativa = np.delete(embeddingsNarrativa, index, axis=0)
              embeddingsModelo = np.delete(embeddingsModelo, j, axis=0)
              # Recuando um índice
              index -= 1
              break
        indexTermos += 1
    for j, frase_modelo in enumerate(parteModelo):
      #print(frase_modelo)
      # Utilizamos o transformer para verificar se são parecidas
      embeddingModelo = embeddingsModelo[j]
      # Vê a similaridade entre os dois dados
      similaridade = util.cos_sim(embeddingNarrativa, embeddingModelo)
      #print(similaridade)
      if similaridade > 0.75:
        #print("\n\n\n\n\n")
        #print(termoTexto, "-------------")
        #print(frase_modelo)
        #print(similaridade)
        #print("\n\n\n\n\n")
        acertos.append(frase_modelo)
        # Removemos os termos
        parteNarrativa.pop(index)
        parteModelo.pop(j)
        # Removemos os termos do embedding
        embeddingsNarrativa = np.delete(embeddingsNarrativa, index, axis=0)
        embeddingsModelo = np.delete(embeddingsModelo, j, axis=0)
        # Recuando um índice
        index -= 1
        break
    index += 1  # Avançando para o próximo termo da narrativa

  return acertos, parteNarrativa, parteModelo
```

```python
def Automatiza(Narrativa, Saida, texto):
  # Separarmos em partes de sintomas, sindromes e diagnóstico a narrativa é a saída do modelo
  sintomasNarrativa, sindromesNarrativa, dadosNegados = separarDados(Narrativa)
  sintomasModelo, sindromesModelo, dadosNegadosModelo = separarDados(Saida)

  # Depois removemos os sintomas negados encontrados na Narrativa
  sintomasNarrativa, sindromesNarrativa = removerDados(sintomasNarrativa, sindromesNarrativa, dadosNegados)

  # A partir dessa parte podemos ver o que o modelo acertou totalmente e parcialmente, assim como o que ele errou e faltou
  # Vamos focar nos acertos totais por enquanto

  # Listas para retornar o resultado final
  acertosTotais = []
  acertosParciais = []
  achadosFaltantes = []
  achadosErrados = []
  achadosNegadosCorretos = []
  achadosNegadosFaltantes = []
  achadosNegadosErrados = []

  # Verificamos os acertos totais usando a biblioteca jellfish
  # Para a função de acertosCompletos passamos a lista de todos os acertos, assim como as duas respectivas listas que queremos ver
  # Realizamos isso pois assim podemos ter uma lista com todos os acertos completos e dentro da função removemos os acertos completos das duas listas para conseguir ver o que vai sobrar em cada
  acertosSintomas, sintomasNarrativa, sintomasModelo = compararDados(sintomasNarrativa, sintomasModelo)
  acertosTotais.append(acertosSintomas)
  # Sindromes
  acertosSindromes, sindromesNarrativa, sindromesModelo = compararDados(sindromesNarrativa, sindromesModelo)
  acertosTotais.append(acertosSindromes)

  # Aqui em baixo serve para os acertos parciais
  # Isso é, ele acertou um dado mas colocou ele na parte errada
  # Por exemplo, podemos ver se um sintoma foi colocado como sindrome ou diagnostico
  parcialSintoma, sintomasNarrativa, sindromesModelo = acertosParcias(sintomasNarrativa, sindromesModelo)
  acertosParciais.append(parcialSintoma)
  # Acerto parcial de sindromes
  parcialSindrome, sindromesNarrativa, sintomasModelo = acertosParcias(sindromesNarrativa, sintomasModelo)
  acertosParciais.append(parcialSindrome)

  # Com todos as informações retiradas, vamos ver se as negações colocadas pelo modelo estão certas ou não
  acertosNegados, dadosNegados, dadosNegadosModelo = removerDadosNegados(dadosNegados, dadosNegadosModelo)
  achadosNegadosCorretos.append(acertosNegados)

  # Agora que possuimos os acertos, vamos ver sobre o que faltou ou o que o modelo errou
  achadosFaltantes.append(sintomasNarrativa)
  achadosFaltantes.append(sindromesNarrativa)
  achadosNegadosFaltantes.append(dadosNegados)

  achadosErrados.append(sintomasModelo)
  achadosErrados.append(sindromesModelo)
  achadosNegadosErrados.append(dadosNegadosModelo)

  # Muitas vezes o algoritmo identifica um espaço vazio como algo, por isso vamos limpar tudo agora
  acertosTotais = removerDadosVazios(acertosTotais)
  acertosParciais = removerDadosVazios(acertosParciais)
  achadosFaltantes = removerDadosVazios(achadosFaltantes)
  achadosErrados = removerDadosVazios(achadosErrados)

  achadosNegadosCorretos = removerDadosVazios(achadosNegadosCorretos)
  achadosNegadosFaltantes = removerDadosVazios(achadosNegadosFaltantes)
  achadosNegadosErrados = removerDadosVazios(achadosNegadosErrados)

  # Aqui mostramos as coisas para o úsuario
  print("O que está certo completamente: ", acertosTotais)
  print("O que foi parcialmente acertado: ", acertosParciais)
  print("O que o modelo faltou de mencionar: ", achadosFaltantes)
  print("O que o modelo errou ou inventou: ", achadosErrados)
  print("Os dados negados acertados: ", achadosNegadosCorretos)
  print("Os dados negados que faltou de mencionar: ", achadosNegadosFaltantes)
  print("Os dados negados errados: ", achadosNegadosErrados, "\n")


  # Aqui é a parte final do código, onde colocamos os dados na posição correta da coluna
  escreverDados(acertosTotais, respostaColuna + 2, file_name, texto)
  escreverDados(acertosParciais, respostaColuna + 3, file_name, texto)
  escreverDados(achadosFaltantes, respostaColuna + 4, file_name, texto)
  escreverDados(achadosErrados, respostaColuna + 5, file_name, texto)
  escreverDadosNegados(achadosNegadosCorretos, respostaColuna + 6, file_name, texto)
  escreverDadosNegados(achadosNegadosFaltantes, respostaColuna + 7, file_name, texto)
  escreverDadosNegados(achadosNegadosErrados, respostaColuna + 8, file_name, texto)

  return 0
```

Funções utilizada para a avaliação de como o modelo se saiu

```python
def Avaliacao(acertosCompletos, acertosParciais, acertosFaltantes, acertosErrados, dadosNegadosAcertados, dadosNegadosFaltantes, dadosNegadosErrados):
    # Com esses dados podemos calcular e extrair algumas informações interessantes
    # A Acurácia do modelo
    # Definindo a variável de ajuste para achados parciais
    alpha = 0.5

    # Cálculo da revocação ajustada
    denominador_recall = acertosCompletos + acertosParciais + acertosFaltantes + dadosNegadosAcertados + dadosNegadosFaltantes
    if denominador_recall != 0:
        recall = (acertosCompletos + (acertosParciais * alpha) + dadosNegadosAcertados) / denominador_recall
    else:
        recall = 0

    # Cálculo da precisão ajustada
    denominador_precisao = acertosCompletos + acertosParciais + acertosErrados + dadosNegadosAcertados + dadosNegadosErrados
    if denominador_precisao != 0:
        precisao = (acertosCompletos + (acertosParciais * alpha) + dadosNegadosAcertados) / denominador_precisao
    else:
        precisao = 0

    # O f-score do modelo
    if precisao != 0 or recall != 0:
        fScore = 2 * ((precisao * recall) / (precisao + recall))
    else:
        fScore = 0

    # Mostrando para o usuário
    #print("Precisão: ", precisao)
    #print("Recall: ", recall)
    #print("F-Score: ", fScore)

    return precisao, recall, fScore
```
